```python
# Excercise 7
def count_words(text):
    # Convert to lowercase and remove punctuation
    text = text.lower()

    for char in ".,!?;:'\"()[]{}":
        text = text.replace(char, "")

    words = text.split()

    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

    return word_count
        
```

```python
# excercise 9

def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result
```

```python
# Excercise 10

def analyze_student_scores(students):
    # Calculate average score for each student
    averages = {}

    for student in students:
        avg_score = sum(student["scores"]) / len(student["scores"])
        averages[student["name"]] = avg_score

    # Identify the student with the highest average
    top_student = max(averages, key=averages.get)

    return {
        "averages": averages,
        "top_student": {
            "name": top_student,
            "average_score": averages[top_student]
        }
    }
```

```python
# Final Challenge

# You're given a list of dictionaries representing monthly sales data for a store.
# Each dictionary contains the month, category, and sales amount.

sales_data = [
    {"month": "Jan", "category": "Electronics", "amount": 1200},
    {"month": "Jan", "category": "Books", "amount": 850},
    {"month": "Jan", "category": "Clothing", "amount": 980},
    {"month": "Feb", "category": "Electronics", "amount": 1450},
    {"month": "Feb", "category": "Books", "amount": 920},
    {"month": "Feb", "category": "Clothing", "amount": 1050},
    {"month": "Mar", "category": "Electronics", "amount": 1600},
    {"month": "Mar", "category": "Books", "amount": 780},
    {"month": "Mar", "category": "Clothing", "amount": 1200}
]

def analyze_sales(sales):
    # Step 1 Calculate total sales per month
    sales_per_month = {}
    for record in sales:
        month = record["month"]
        sales_per_month[month] = sales_per_month.get(month, 0) + record["amount"]

    # Step 2: Calculate total sales per category
    sales_per_category = {}
    for record in sales:
        category = record["category"]
        sales_per_category[category] = (
            sales_per_category.get(category, 0) + record["amount"]
        )

    # Step 3: Find the best performing month and category
    best_month = max(sales_per_month, key=sales_per_month.get)
    best_category = max(sales_per_category, key=sales_per_category.get)
    
    # Step 4: Calculate overall total sales
    overall_total = sum(record["amount"] for record in sales)
    
    return {
        "sales_per_month": sales_per_month,
        "sales_per_category": sales_per_category,
        "best_month": (best_month, sales_per_month[best_month]),
        "best_category": (best_category, sales_per_category[best_category]),
        "overall_total": overall_total,
    }
    
```

```python
# Test Challenge

result = analyze_sales(sales_data)

for key, value in result.items():
    print(f"{key}: {value}")
```
