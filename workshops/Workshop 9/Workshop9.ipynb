```python
import pandas as pd
customers = pd.read_csv('customers_wk9_thurs.csv')
products = pd.read_csv('products_wk9_thurs.csv')
sales = pd.read_csv('sales_wk9_thurs.csv')
```

```python
# Convert date columns

customers['registration_date'] = pd.to_datetime(customers['registration_date'])
products['launch_date'] = pd.to_datetime(products['launch_date'])
sales['transaction_date'] = pd.to_datetime(sales['transaction_date'])
```

```python
# Extract Year, Month, and Day of the Week

sales['year'] = sales['transaction_date'].dt.year
sales['month'] = sales['transaction_date'].dt.month_name()
sales['day_of_week'] = sales['transaction_date'].dt.day_name()
```

```python
# Display the sales date range
print('Earliest Sale:', sales['transaction_date'].min())
print('Latest Sale:', sales['transaction_date'].max())
```

```python
# Basic Time Analysis
## Busiest day of the week
print(sales['day_of_week'].value_counts())
```

```python
## Month with the most transaction
print(sales['month'].value_counts())
```

```python
## Sales by year

print(sales['year'].value_counts().sort_index())
```

```python
# Part 2 GroupBy and Aggregation
```

```python
# Product Analysis
## Average price by category
print(products.groupby('category')['price'].mean())
```

```python
## Number of products by category

print(products.groupby('category').size())
```

```python
# Customer Analysis
## Average age by state
print(customers.groupby('state')['age'].mean())
```

```python
## Total spending per customer
customer_spending = sales.groupby('customer_id')['total_amount'].sum()
print(customer_spending)
```

```python
## Customers with the most purchases

purchase_count = sales.groupby('customer_id').size().sort_values(ascending=False)
print(purchase_count)
```

```python
# Sales Analysis
## Total sales revenue by month

sales['month_period'] = sales['transaction_date'].dt.to_period('M')
monthly_sales = sales.groupby('month_period')['total_amount'].sum()
print(monthly_sales)
```

```python
## Average transaction amount by day of the week

avg_day = sales.groupby('day_of_week')['total_amount'].mean()
print (avg_day)
```

```python
## Product category with the most revenue

sales_products = sales.merge(products[['product_id','category']], on='product_id')
category_revenue = sales_products.groupby('category')['total_amount'].sum()
print(category_revenue.sort_values(ascending=False))
```

```python
# Part 3 Basic Data Insights
## State with the most customers

print(customers['state'].value_counts())
```

```python
## Top 5 customers by spending

top5 = sales.groupby('customer_id')['total_amount'].sum().sort_values(ascending=False)
print(top5.head())
```
